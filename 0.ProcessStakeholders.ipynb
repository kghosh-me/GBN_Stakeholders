```python
import pandas as pd
import os, glob

from owlready2 import *
import owlready2
print(owlready2.VERSION)

import json, requests
import OAI
```

```python
h = OAI.Helper("stakeholders_NOA","./cache")
```

## Loading stakeholders

```python
onto = owlready2.get_ontology("data/WIP.owl").load()
sg = (onto.StakeholderGroup.instances())
ssg = (onto.StakeholderSubgroup.instances())
ssg_s = {}
for g in ssg:
    ssg_s[g] = g.has_SubgroupStakeholder
```

* "Needs Opportunities Abilities" framework
* capability (C), opportunity (O), and motivation (M) 

```python
CAT = {}
for SSG in list(ssg):
    SG = SSG.has_SubgroupStakeholder[0].has_StakeholderGroup[0]
    if SG not in CAT.keys():
        CAT[SG] = []
    CAT[SG].append(SSG)
    CAT[SG] = list(set(CAT[SG]))
    pr  = """You are an expert stakeholders management, in particular cities stakeholder. You study in particular their behaviors motivations, needs, and capabilities in green cities, and smart and sustainable neighbourhoods.
    You are now reviewing: """
    pr += "'"+SSG.label[0]+"', which are described as follows:\n" + SSG.has_Description[0].split("\n\nSources: ")[0].strip()
    N = h.ask(pr,"Considering the NOA (Needs, Opportunities, Abilities) framework, how would you define their 'needs' (or motivation) - both automatic and reflective - in the context of green cities (also known as GBN, Green Building neighbourhoods) ?",ow=False)
    O = h.ask(pr,"Considering the NOA (Needs, Opportunities, Abilities) framework, how would you define their 'opportunities' - both physical and social - in the context of green cities (also known as GBN, Green Building neighbourhoods) ?",ow=False)
    A = h.ask(pr,"Considering the NOA (Needs, Opportunities, Abilities) framework, how would you define their 'abilities' (or capabilities) - both physical and psychological - in the context of green cities (also known as GBN, Green Building neighbourhoods) ?",ow=False)

    pr2 = """You are an expert stakeholders management, in particular cities stakeholder. You study in particular their behaviors motivations, needs, and capabilities in green cities, and smart and sustainable neighbourhoods.
    You are now reviewing: """
    pr2 += "'"+SSG.label[0]+"', which are described as follows:\n" + SSG.has_Description[0].split("\n\nSources: ")[0].strip()+"\n\nTheir NOA (Needs, Opportunities, Abilities) assessment yielded the following:\n\n"
    pr2 += "### Needs\n\n"+N+"\n\n"
    pr2 += "### Opportunities\n\n"+O+"\n\n"
    pr2 += "### Capabilities\n\n"+A+"\n\n"

    INTRO = h.ask(pr2,"In a paragraph, can you summarize how their behaviors are driven by their Needs, Opportunities and Capabilities? Be as specific as possible.",ow=False)

    MD  = "# "+SSG.label[0]+"\n\n"
    MD += "## Description\n\n"+SSG.has_Description[0].split("\n\nSources: ")[0].strip()+"\n\n"
    MD += "## NOA\n\n"
    MD += "### Overview\n\n"+INTRO+"\n\n"
    MD += "### Needs\n\n"+N+"\n\n"
    MD += "### Opportunities\n\n"+O+"\n\n"
    MD += "### Capabilities\n\n"+A+"\n\n"

    

    MD += "## Members of this group\n\n"
    MD += "* "+"\n* ".join([x.label[0] for x in SSG.has_SubgroupStakeholder])
    with open("pages/"+SSG.name+".md","w") as f:
        f.write(MD)
```

# Producing the index

```python
MD  = "# NOA review of stakeholders categories from T3.5\n\n"
MD += "Disclaimer: this is just a quick and dirty proof of concept for automating stuff\n\n"
for category in CAT.keys():
    MD += "\n\n## "+category.label[0]+"\n\n"
    MD += "* "+"\n* ".join([ "["+x.label[0]+"]("+x.name+".md)" for x in CAT[category]])
with open("pages/Readme.md","w") as f:
        f.write(MD)
```
